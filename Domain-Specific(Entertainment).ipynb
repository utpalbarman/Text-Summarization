```python
# Data Cleaning(Final) and Preprocessing, where lowercasing, removal of bracketed text, quotation marks, contractions conversion,
#possessive noun conversion, special characters and punctuations removel with stop words removal after lemmatization and joined
def preprocess(text):
    lem = WordNetLemmatizer()
    new_string = text.lower()
    new_string = re.sub(r'\([^)]*\)', '', new_string)
    new_string = re.sub('"','', new_string)
    new_string = ' '.join([contraction_mapping[t] if t in contraction_mapping else t for t in new_string.split(" ")]) 
    new_string = re.sub(r"'s\b","",new_string)
    new_string = re.sub("[^a-zA-Z]", " ", new_string) 
    new_string = [lem.lemmatize(w) for w in new_string.split() if not w in stop_words]
    new_string = " ".join(new_string)
    return new_string
```

```python
def clean(text):
    cleaned = []
    for y in text:
        cleaned.append(preprocess(y))
    return cleaned
```

```python
documents2=documents1.apply(clean)
summaries2=summaries1.apply(clean)
```

```python
#Leveraging Static Word Embeddings to create vector reporesentation of the words
vec_path = './glove.6B/glove.6B.100d.txt'
embeddings_file = open(vec_path, encoding="utf-8")
word_embeddings = {}
for line in embeddings_file:
    word, coefs = line.split(maxsplit=1)
    coefs = np.fromstring(coefs, "f", sep=" ")
    word_embeddings[word] = coefs
```

```python

def glove(cleaned_text):
    glove_vectors = []
    for i in cleaned_text:
        if len(i)!=0:
            v = sum([word_embeddings.get(w, np.zeros((100,))) for w in i.split()])/(len(i.split())+0.001)
        else:
            v = np.zeros((100,))
        glove_vectors.append(v)
    return glove_vectors
```

```python
document_vectors = documents2.apply(glove)
summary_vectors = summaries2.apply(glove)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import networkx as nx
from rouge import Rouge
```

```python
#Initialize Empty Square Matrix
sim_mat=[]
for i in document_vectors:
    sim_mat.append(np.zeros([len(i), len(i)]))
```

```python
#Cosine Similarity between each and every sentences
start_time = time.time()
i=-1
for z in document_vectors:
    i+=1
    for y in range(len(z)):
        for x in range(len(z)):
            
            if y!=x:
                sim_mat[i][y][x] = cosine_similarity(z[y].reshape(1,100), z[x].reshape(1,100))[0,0]
```

```python
#Sentence ranking
def rank(text, scores):
    ranked = []
    for x in range(len(text)):
        ranked_sentences = sorted(((scores[x][i],s) for i,s in enumerate(text[x])), reverse=True)
        ranked.append(ranked_sentences)
    return ranked
```

```python
#Sentence Extraction
def summarize(ranked_sentences, length):
    summaries = []
    for sentences in ranked_sentences:
        no = int((length * len(sentences))/100)
        summary = ""
        if len(sentences)>no:
            for i in range(no):
                summary += sentences[i][1] + " "
        else:
            for i in range(len(sentences)):
                summary += sentences[i][1] + " "
        summaries.append(summary)
    return summaries
```

```python
#ROUGE Evalaation
def rouge_2FMeasure(hypothesis,reference):
    rouge = Rouge()
    scores = []
    for i in range(len(hypothesis)): 
        score = rouge.get_scores(hypothesis[i], reference[i])[0]['rouge-1']['f']
        scores.append(score)
    return score
rouge_2FMeasure(textrank_summaries,df_articles['summary'])
```

```python
def rouge_all(hypothesis, reference):
    rouge=Rouge()
    for i in range(len(hypothesis)): 
        score = rouge.get_scores(hypothesis[i], reference[i],avg=True)
        rouge1f = score['rouge-1']['f']
        rouge1p = score['rouge-1']['p']
        rouge1r = score['rouge-1']['r']
        rouge2f = score['rouge-2']['f']
        rouge2p = score['rouge-2']['p']
        rouge2r = score['rouge-2']['r']
        rougelf = score['rouge-l']['f']
        rougelp = score['rouge-l']['p']
        rougelr = score['rouge-l']['r']
        keypoint = score.keys()
        data1=list(score.values())
    return (rouge1f, rouge1p, rouge1r), (rouge2f, rouge2p, rouge2r), (rougelf, rougelp, rougelr), keypoint, data1
```

```python
tot_prec1=0
tot_rec1=0
tot_f1=0
tot_prec2=0
tot_rec2=0
tot_f2=0
tot_precl=0
tot_recl=0
tot_fl=0
for sc in [30,40,50,60,70,80]: 
   
    textrank_summaries1 = summarize(textranked, sc)
    (f11, precision1, recall1), (f12, precision2, recall2), (f1l, precisionl, recalll), keypoint, data1  = rouge_all(textrank_summaries1,df_articles['summary'])
   
    print("\n\nFor sentence count: ", sc)
    
    print(f"\n\nROUGE-1 Scores\nPrecision: {precision1}\nRecall: {recall1}\nF-1 Score: {f11}\n\n")
    print(f"ROUGE-2 Scores\nPrecision: {precision2}\nRecall: {recall2}\nF-1 Score: {f12}\n\n")
    print(f"ROUGE-L Scores\nPrecision: {precisionl}\nRecall: {recalll}\nF-1 Score: {f1l}")
    tot_prec1=tot_prec1 + precision1
    tot_rec1=tot_rec1 + recall1
    tot_f1=tot_f1 + f11
    tot_prec2=tot_prec2 + precision2
    tot_rec2=tot_rec2 + recall2
    tot_f2=tot_f2 + f12
    tot_precl=tot_precl+precisionl
    tot_recl=tot_recl+recalll
    tot_fl=tot_fl+f1l
    df = pd.DataFrame(index = keypoint,columns=['f', 'p', 'r'], data=data1)
    df.rename(columns = {'f':'F-Measure', 'p':'Precision','r':'Recall'}, inplace = True)
    
    plt.rcParams['figure.figsize'] = [10,6]
    df.plot.bar()
    plt.title("TextRank Algorithm ROUGE score for BBC News Dataset")
print("Average Rouge 1 Precision:" ,tot_prec1/6)
print("\nAverage Rouge 1 Recall:" ,tot_rec1/6)
print("\nAverage Rouge 1 F-Measure:", tot_f1/6)
print("\nAverage Rouge 2 Precision:", tot_prec2/6)
print("\nAverage Rouge 2 Recall:" ,tot_rec2/6)
print("\nAverage Rouge 2 F-Measure:" ,tot_f2/6)
print("\nAverage Rouge L Precision:", tot_precl/6)
print("\nAverage Rouge L Recall:" ,tot_recl/6)
print("\nAverage Rouge L F-Measure:" ,tot_fl/6)
```
