```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
font = {'size'   : 16}
matplotlib.rc('font', **font)
matplotlib.rcParams['font.sans-serif']= "Arial"
matplotlib.rcParams['font.family'] = "sans-serif"
```

### Domain Agnostic

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-pastel') 
#Evaluation Graph
Precision= [0.59337,0.58147,0.57695]
Recall= [0.76491,0.63526,0.80226]
F_Measure= [0.65755,0.58693,0.66284]
index=['TextRank', 'LSA', 'TF-IDF']
df1 = pd.DataFrame({'Recall': Recall,'Precision': Precision, 'F-measure': F_Measure}, index=index)
```

```python
ax1=df1.plot.bar(rot=0)
ax1.set_xlabel("Algorithms",fontsize=16,labelpad=20)
ax1.set_ylabel("ROUGE-2 Scores",fontsize=16,labelpad=20)
ax1.legend(ncol=5, loc='upper center',bbox_to_anchor=(0.5, 1.0))
plt.ylim(0, 1.0)
plt.rcParams['figure.figsize'] = [12, 8]
plt.title("ROUGE-2 Scores of each algorithms in BBC News(Domain-Agnostic)",pad=20)
plt.savefig('Domain Specific(Domain Agnostic1).tiff',dpi=600)
```
